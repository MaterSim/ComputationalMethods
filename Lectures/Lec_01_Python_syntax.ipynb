# Exercise
In 1888 Johannes Rydberg published his famous formula for the wavelengths $\lambda$ of the emissioon lines of the hydrogen atom


$$\frac{1}{\lambda} = R (\frac{1}{m^2} - \frac{1}{n^2})$$

Where $R$ is the Rydberg constant = 1.097 $\times$ $10^{-2} \text{nm}^{-1}$, $m$ and $n$ are positive integers. For a given value of $m$, $\lambda$ is given for this formula for all $n>m$ series, the first three series are known as the Lyman, Belmer and Paschen series, named after their discoverers. Can you write a code to compute the first three series for the first five lines?

````python
R = 1.097e-2
for m in [1,2,3]:
    print("Series for m = ", m)
    for k in [1,2,3,4,5]:
        n = m + k
        invlambda = R*(1.0/m**2 - 1/n**2)
        print("n=", n, "  ", 1/invlambda, "nm")
````

In general, there will be two mechanisms to skip
- break, which completely exits from the given for/while loops
- continue, skip the current loop, but will continue to carry on

This continues the rest of the loop. Sometimes when a condition is satisfied there are chances of the loop getting terminated. This can be avoided using continue statement.

## Final Tips before starting your Python programming

1. Be aware that you must end up with : sign, when you write the sentence of condition

2. Indentation is required by python

````python
print("\nUsing 'continue':")
for i in range(5):
  if i == 2:
    print("Skipping iteration 2 with 'continue'")
    continue
  print(f"Current value of i (continue): {i}")

print("\nUsing 'break':")
for i in range(5):
  if i == 2:
    print("Breaking loop at iteration 2 with 'break'")
    break
  print(f"Current value of i (break): {i}")
````
